```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("generatorv8.csv")
```

```python
df.head()
```

```python
import matplotlib
matplotlib.get_cachedir()
```

```python
x = range(1, 60)
plt.plot(x, df['disc_loss'], label="Discriminator Loss")
plt.plot(x, df['gen_loss'], label="Generator Loss")
plt.plot(x, df['val_loss'], label="Validation Loss")
plt.plot(x, np.zeros(59), linestyle="dashed", label="Desired Value")
leg = plt.legend(loc="lower right")

plt.title("cGAN Loss Over 60 Epochs")
plt.xlabel("Epoch #")
plt.ylabel("Loss (Wasserstein Distance)")
sns.set_theme()

plt.show('darkgrid')
```
