```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
train = pd.read_csv("./ppo1/logtrainparse.csv")
test = pd.read_csv("./ppo1/testlogparse.csv")
```

```python
train.head()
```

```python
x = range(1, 69)
plt.plot(x, train["ep_rew_mean"], label="Reward")
#plt.plot(x, train['loss'], label="Network Loss")
plt.xlabel("Batch Index")
plt.ylabel("Reward (PPO)")
plt.title("Navigation Model V1 Reward Over Batches")
leg = plt.legend(loc="upper left")
plt.xlim([0.75, 69])
sns.set_theme()
plt.show('darkgrid')
```

```python
x = range(1, 69)
#plt.plot(x, train["ep_rew_mean"], label="Reward")
plt.plot(x, train['loss'], label="Network Loss")
plt.xlabel("Batch Index")
plt.ylabel("Loss (Surrogate PPO Loss with KL & Clip)")
plt.title("Navigation Model V1 Loss Over Batches")
leg = plt.legend(loc="upper left")
plt.xlim([0.75, 69])
sns.set_theme()
plt.show('darkgrid')
```

```python
x = range(1, 16)
#plt.plot(x, train["ep_rew_mean"], label="Reward")
plt.plot(x, test['reward'], label="Testing Reward")
plt.xlabel("Batch Index")
plt.ylabel("Reward (PPO)")
plt.title("Navigation Model V1 Reward During Testing")
leg = plt.legend(loc="upper right")
plt.xlim([0.75, 16])
sns.set_theme()
plt.show('darkgrid')
```
